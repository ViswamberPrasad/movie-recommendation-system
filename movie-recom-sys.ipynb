WE CONVERT TEXTS TO VECTORS THROUGH A PROCESS CALLED TEXT VECTORISATION, IN PARTICULAR USING A METHOD CALLED "BAG OF WORDS"

CONVERTING THE TAGS INTO 5000 MOST COMMON WORDS AND FILLING A DATAFRAM AGAINST 5000 MOVIES, SO WE GET A VECTORISED FORM OF EACH MOVIES

https://scikit-learn.org/stable/modules/generated/sklearn.feature_extraction.text.CountVectorizer.html

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
#the cv object is an instance of the Count Vectorizer class and extracts top 5000 words and ignores all the stop words present in english language
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
#.fit_transform method of the Count Vectoriser class, fits the tags column to extract 5000 features eliminating stop words
#fit_transform() method returns a sci_py sparse matrix so we convert it to a numpy array to see the values and store in vectors
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
ps.stem('loving')
```

```python
#defining a helper function to stem the corpus
def stem(txt):
    y = [ps.stem(x) for x in txt.split()]
    return " ".join(y)    
```

```python
#we now have movies as vectors in a 5000 dimensional space, so now we have to have to calculate the distance of one movie
#from another to measure the similarity.
#We dont use euclidean distance since its a larger space, we rather use cosine distance.
```

https://scikit-learn.org/stable/modules/generated/sklearn.metrics.pairwise.cosine_similarity.html

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
type(similarity)
```

```python
#we now need to sort the array of movies in such a way that the 
#more similar movies from a particular movie are on top
#so we can recommend the top 5 least distance -> most similar movies
```

```python
sorted(similarity[0],reverse = True) #in this way we can sort in descending order, but we loose the index values of movies
```

```python
#enumerate adds index values to the iterators or the objects, hence we dont loose it
list(enumerate(similarity[0]))
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
new_df[new_df['title'] == 'Batman Begins'].index[0] #The index property returns the index information of the DataFrame.
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    movie_fetch = similarity[movie_index]
    movies_list = sorted(list(enumerate(movie_fetch)),reverse=True,key=lambda x:x[1])[1:6]
    print([new_df.iloc[x[0]].title for x in movies_list])
```
